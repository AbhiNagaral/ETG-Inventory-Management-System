# Adding in inventory

```python
import json
```

```python
fd=open('Record.json','r')
r=fd.read()
fd.close()

```

```python
record=json.loads(r)
```

```python
record
```

```python
prod_id=input("Enter product id:")
if prod_id in record:
    print("Product id already present. Update quantity to be added")
    qty=int(input("Enter quantity: "))
    qty+=record[prod_id]['qty']
else:
    prod_name=input("Enter product name: ")
    qty=int(input("Enter quantity: "))
    price=int(input("Enter price: "))
    disc=int(input("Enter discount: "))
    val=int(input("Enter validity of product: "))

    record[prod_id]={'name':prod_name,'price':price,'qty':qty,'discount':disc,'validity':val}
```

```python
record
```

```python
new= json.dumps(record)
```

```python
fd=open('Record.json','w')
fd.write(new)
fd.close()
```
